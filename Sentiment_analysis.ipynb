## 14. Market Sentiment Strength vs Profit
>Correlation: Fear & Greed Index vs Profit


```python
import numpy as np

x = sentimental_analysis_df["value"]
y = sentimental_analysis_df["Closed PnL"]

plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(x, y, alpha=0.5)

# Trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.plot(x, p(x))

plt.title("Fear & Greed Index vs Profit")
plt.xlabel("value")
plt.ylabel("Closed PnL")

plt.tight_layout()
plt.show()
```

### Insight

- The scatter plot shows the relationship between **Fear & Greed Index (value)** and **Closed PnL (profit)**.
- Each point represents a trade, and the trendline helps identify the overall direction.

- If the trendline is **positive**, it suggests that higher Fear & Greed values (Greed conditions) are associated with higher profits.
- If the trendline is **negative**, it suggests that higher Fear & Greed values are associated with lower profits.

- The scattered distribution of points indicates that profit is **not strongly dependent only on sentiment value**, and other factors (coin, trade direction, timing) also influence performance.

- Overall, sentiment has an impact, but it is **not a perfect predictor of profit**.
