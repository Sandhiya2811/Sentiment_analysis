### 6. No Strong Negative Correlations Found
- Most correlations are weak or positive.
- There is no evidence of a strong inverse relationship between the analyzed trading metrics.

### Key Business Insight
- Trade size is the most important factor influencing trader performance. Larger trades are associated with higher profits, but they also lead to higher trading fees. Traders should focus on position sizing strategies to maximize profitability while controlling transaction costs.

# Does high trade size increase profit?

```python
size_profit = sentimental_analysis_df.copy()

size_profit['Size Bucket'] = pd.qcut(
    size_profit['Size USD'],
    5,
    labels=["Very Low", "Low", "Medium", "High", "Very High"]
)

size_analysis = (
    size_profit
    .groupby('Size Bucket')['Closed PnL']
    .mean()
    .reset_index()
)

```

```python
size_analysis
```

```python
fig = px.bar(
    size_analysis,
    x="Size Bucket",
    y="Closed PnL",
    title="Trade Size vs Profit",
    text="Closed PnL",
    color="Closed PnL",
    color_continuous_scale="Viridis"
)

fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.show()
```

## Profitability by Trade Size

### 1. Profit Increases with Trade Size
- Average Closed PnL increases as trade size grows.
- Larger trades tend to generate higher profits than smaller trades.

### 2. Very Low Trade Sizes Generate Minimal Profit
- Very Low trade sizes recorded an average Closed PnL of only 1.68.
- Small positions contributed the least to overall profitability.

### 3. Medium and High Trade Sizes Show Strong Performance
- Medium-sized trades generated 73.75 in average profit.
- High-sized trades generated 91.98 in average profit.
- Increasing position size appears to improve profit potential.

### 4. Very High Trade Sizes Deliver the Highest Profit
- Very High trade sizes recorded the highest average Closed PnL of 252.86.
- These trades generated nearly three times the profit of High-sized trades.

### 5. Clear Positive Relationship Between Trade Size and Profit
- Profitability consistently increases from Very Low to Very High trade sizes.
